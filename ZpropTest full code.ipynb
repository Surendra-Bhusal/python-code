# **FORMULA FOR THE Z STATISTICS.**

**z = (p-p0) / âˆš(p0(1-p0)/n)**

p= sample proportion, p0= population proportion , n = sample size


 **Let's first deal with denominator part âˆš(p0(1-p0)/n**     
 which is also known as margin of error.

```python
MOE= np.sqrt((0.9*(1-0.9))/100)
MOE
```

**Number of patient in first sample receiving aspirin**

```python
First_Sample_of_100_people[First_Sample_of_100_people['Receives aspirin']=='Yes'].shape[0]
```

**Number of patient in second sample receiving aspirin**

```python
Second_Sample_of_100_people[Second_Sample_of_100_people['Receives aspirin']=='Yes'].shape[0]
```

**Proportion of patient in first sample receiving aspirin**

```python
First_Sample_of_100_people[First_Sample_of_100_people['Receives aspirin']=='Yes'].shape[0]/100
```

**Proportion of patient in second sample receiving aspirin**

```python
Second_Sample_of_100_people[Second_Sample_of_100_people['Receives aspirin']=='Yes'].shape[0]/100
```

**SO TO CALCULATE Z STATISTICS FOR THE FIRST SAMPLE**

**WE CAN SIMILARLY CALCULATE THE Z STATISTICS FOR 100 RANDOMLY SELECTED SAMPLE AND STORE THEM IN A LIST**

```python
list1= []
for i in range (0,100):
    new_df = df.sample(n=100)
    z= (((new_df[new_df['Receives aspirin']=='Yes'].shape[0])/100)- 0.90 )/MOE
    list1.append(z)

```

```python
list1
```

**WE CAN PLOT THE VALUES OF Z STATISTICS AS HISTOGRAM**

```python
# prompt: plot the values from list1 as histogram

import matplotlib.pyplot as plt

plt.hist(list1)
plt.xlabel('Z-scores')
plt.ylabel('Frequency')
plt.title('Histogram of Z-scores')
plt.show()
```

**WE CAN SIMILARLY CALCULATE THE Z STATISTICS FOR 10000 RANDOMLY SELECTED SAMPLE**

**NOW LETS SEE WHAT PROPORTIONS OF Z STATISTICS VALUE ARE BETWEEN AN ARBITRARY VALUE -1.96 AND +1.96**

```python
# prompt: calculate the proportion of numbers between -1.96 and +1.96 in list 3

import numpy as np
number = np.sum((np.array(list3) >= -1.96) & (np.array(list3) <= 1.96))
proportion = np.sum((np.array(list3) >= -1.96) & (np.array(list3) <= 1.96)) / len(list3)
print("Number :", number)
print ("Proportion :" ,proportion)
```
